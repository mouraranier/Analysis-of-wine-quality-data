```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('winequality/winequality-red.csv', sep = ";")
```

# Select four features and then compute: mean, mode, median, percentile (20, 50 and 70), quartile (Q1, Q2 and Q3), variance and standard deviation

```python
df.describe()
```

```python
#mÃ©dia

media_density = np.mean(df["density"])
media_pH = np.mean(df["pH"])
media_residual_sugar = np.mean(df["residual sugar"])
media_chlorides = np.mean(df["chlorides"])

media_density, media_pH, media_residual_sugar, media_chlorides
```

```python
#moda

moda_density = stats.mode(df["density"])
moda_pH = stats.mode(df["pH"])
moda_residual_sugar = stats.mode(df["residual sugar"])
moda_chlorides = stats.mode(df["chlorides"])

moda_density, moda_pH, moda_residual_sugar, moda_chlorides
```

```python
#mediana

mediana_density = np.median(df["density"])
mediana_pH = np.median(df["pH"])
mediana_residual_sugar = np.median(df["residual sugar"])
mediana_chlorides = np.median(df["chlorides"])

mediana_density, mediana_pH, mediana_residual_sugar, mediana_chlorides
```

```python
#percentis 20

percentis20_density = np.percentile(df["density"], 20)
percentis20_pH = np.percentile(df["pH"], 20)
percentis20_residual_sugar = np.percentile(df["residual sugar"], 20)
percentis20_chlorides = np.percentile(df["chlorides"], 20)

percentis20_density, percentis20_pH, percentis20_residual_sugar, percentis20_chlorides

```

```python
#percentis 50
#igual a mediana

percentis50_density = np.percentile(df["density"], 50)
percentis50_pH = np.percentile(df["pH"], 50)
percentis50_residual_sugar = np.percentile(df["residual sugar"], 50)
percentis50_chlorides = np.percentile(df["chlorides"], 50)

percentis50_density, percentis50_pH, percentis50_residual_sugar, percentis50_chlorides
```

```python
#percentis 70

percentis70_density = np.percentile(df["density"], 70)
percentis70_pH = np.percentile(df["pH"], 70)
percentis70_residual_sugar = np.percentile(df["residual sugar"], 70)
percentis70_chlorides = np.percentile(df["chlorides"], 70)

percentis70_density, percentis70_pH, percentis70_residual_sugar, percentis70_chlorides
```

```python
#Quartil Q1
#Quartil Q2 igual a mediana e igual ao percentil 50

quartilQ1_density = np.percentile(df["density"], 25)
quartilQ1_pH = np.percentile(df["pH"], 25)
quartilQ1_residual_sugar = np.percentile(df["residual sugar"], 25)
quartilQ1_chlorides = np.percentile(df["chlorides"], 25)

quartilQ1_density, quartilQ1_pH, quartilQ1_residual_sugar, quartilQ1_chlorides
```

```python
#Quartil Q3

quartilQ3_density = np.percentile(df["density"], 75)
quartilQ3_pH = np.percentile(df["pH"], 75)
quartilQ3_residual_sugar = np.percentile(df["residual sugar"], 75)
quartilQ3_chlorides = np.percentile(df["chlorides"], 75)

quartilQ3_density, quartilQ3_pH, quartilQ3_residual_sugar, quartilQ3_chlorides
```

```python
#variancia

variancia_density = np.var(df["density"])
variancia_pH = np.var(df["pH"])
variancia_residual_sugar = np.var(df["residual sugar"])
variancia_chlorides = np.var(df["chlorides"])

variancia_density, variancia_pH, variancia_residual_sugar, variancia_chlorides
```

```python
#desvio padrÃ£o

desviopadrao_density = np.std(df["density"])
desviopadrao_pH = np.std(df["pH"])
desviopadrao_residual_sugar = np.std(df["residual sugar"])
desviopadrao_chlorides = np.std(df["chlorides"])

desviopadrao_density, desviopadrao_pH, desviopadrao_residual_sugar, desviopadrao_chlorides
```

# Compute the covariance and correlation matrices

```python
matriz_cov = pd.DataFrame(np.cov(df, rowvar = False), 
                          columns = ['fixed acidity','volatile acidity','citric acid','residual sugar',
                                     'chlorides','free sulfur dioxide','total sulfur dioxide','density',
                                     'pH','sulphates','alcohol','quality'],
                         index = ['fixed acidity','volatile acidity','citric acid','residual sugar',
                                     'chlorides','free sulfur dioxide','total sulfur dioxide','density',
                                     'pH','sulphates','alcohol','quality'])

print(matriz_cov.shape)
matriz_cov = matriz_cov.round(4)
matriz_cov
```
